# Taking the features only

```python
x = data.iloc[:, 3:5]
x
```

```python
from sklearn.cluster import KMeans
WCSS = []
for i  in range(1, 11):
  model = KMeans(n_clusters = i)
  model.fit(x)
  WCSS.append(model.inertia_)
```

```python
print (WCSS)
```

```python
plt.plot(range(1, 11), WCSS)
```

```python
model = KMeans(n_clusters = 5)
model.fit(x)
```

```python
y = model.fit_predict(x)
```

```python
x['Groups'] = y
```
